### 2.1 Mixed data types

```python
# Check for mixed types:
for col in df_car.columns.tolist():
  weird = (df_car[[col]].applymap(type) != df_car[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_car[weird]) > 0:
    print (col)
```

```python
# Change mixed-type variables
df_car['driver_gender'] = df_car['driver_gender'].astype('str')
```

```python
# Check for mixed types again:
for col in df_car.columns.tolist():
  weird = (df_car[[col]].applymap(type) != df_car[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_car[weird]) > 0:
    print (col)
```
